```python
import import_ipynb
import json
```

```python
fd = open("record.json",'r')
r = fd.read()
fd.close()

record = json.loads(r)
```

## Adding item

```python
prod_id = str(input("Enter product id:"))
Mdl = str(input("Enter Model name:"))
br = str(input("Enter Brand:"))
colr = str(input("Enter colour:"))
pr = int(input("Enter price:"))
ram = str(input("Enter RAM:"))
stg = str(input("Enter Storage:"))
qn = int(input("Enter quantity:"))

record[prod_id] = {"Model name":Mdl, "Brand": br, "colour":colr, "Price": pr, "RAM": ram, "Storage": stg, "quantity": qn}

js = json.dumps(record)

fd = open("record.json",'w')
fd.write(js)
fd.close()
```
