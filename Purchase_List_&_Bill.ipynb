# Read Inventory

```python
import import_ipynb
import json
import datetime
import ctypes
ch='y'
Product_Id=[]
amount=[]
qty=[]
Md_nm=[]
rate=[]
```

```python
fd = open("record.json",'r')
rc= fd.read()
fd.close()

records = json.loads(rc)
```

```python
records
```

# Purchase


```python
while ch=='y' or ch=='Y':    
        ui_prod  = str(input("Enter the product_Id: "))
        Product_Id.append(ui_prod)
        ui_quant = int(input("Enter the quantity: "))
        
        tempmd=records[ui_prod]['Model name']
        temppr=records[ui_prod]['Price']
        tempamount=0

        print("Product: ", records[ui_prod]['Model name'])
        print("Brand: ", records[ui_prod]['Brand'])
        print("Colour: ", records[ui_prod]['colour'])
        print("RAM: ", records[ui_prod]['RAM'])
        print("Storage: ", records[ui_prod]['Storage'])
        print("Price: ", records[ui_prod]['Price'])
        
        Md_nm.append(tempmd)
        rate.append(temppr)
        qty.append(ui_quant)
        amount.append(ui_quant*temppr)
        records[ui_prod]['qn'] = records[ui_prod]['quantity'] - ui_quant
        ch=input('\nDo you want to add more(y for yes and n for no): ')
```

# BILL

```python
d_date = datetime.datetime.now()
reg_format_date = d_date.strftime("  %d-%m-%Y\t\t\t\t\t\t  Bill \t\t\t\t\t  %I:%M:%S %p")
print ('+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++')
print (reg_format_date)
print ('+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++')

print ('\nItemname\t'+'Model\t\t\t'+'\tRate\t'+'Qty\t'+'Amount')
for i in range(0,len(Product_Id)):
    print ('{0}\t{1}\t\t\t\t{2}\t{3}\t{4}'.format(Product_Id[i],Md_nm[i],rate[i],qty[i],amount[i]))
print ('\nTotal amount:\t\t\t\t\t\t\t{0}'.format(sum(amount)))
print('Thank you for shopping here!')
```
